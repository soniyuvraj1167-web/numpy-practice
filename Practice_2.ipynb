```python
import numpy as np
```

```python
# Problem: Add two arrays a and b element-wise.
# Input:
# a = np.array([10, 20, 30, 40])
# b = np.array([1, 2, 3, 4])
# Expected Output: [11 22 33 44]

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
a+b
```

```python
# Problem: Multiply two arrays a and b element-wise.
# Input:
# a = np.array([2, 4, 6])
# b = np.array([3, 5, 7])
# Expected Output: [ 6 20 42]

a = np.array([2, 4, 6])
b = np.array([3, 5, 7])
a*b
```

```python
# Problem: Compute the square of each element in array a.
# Input: a = np.array([1, 2, 3, 4, 5])
# Expected Output: [ 1 4 9 16 25]

a = np.array([1, 2, 3, 4, 5])
a**2
```

```python
# Problem: Compute the square root of each element in array a.
# Input: a = np.array([1, 4, 9, 16, 25])
# Expected Output: [1. 2. 3. 4. 5.]

a = np.array([1, 4, 9, 16, 25])
# a**0.5  or
np.sqrt(a)
```

```python
# Problem: Compute the natural logarithm (base $e$) of array a
# Input: a = np.array([1, np.e, np.e**2])
# Expected Output: [0. 1. 2.]

a = np.array([1, np.e, np.e**2])
np.log(a)
```

```python
# Problem: Calculate the dot product of two 1D vectors.
# a = np.array([1, 2, 3])
# b = np.array([4, 5, 6])
# Expected Output: 32  (Calculation: 1*4 + 2*5 + 3*6)


a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
np.dot(a,b)
```

```python
# Problem: Perform matrix multiplication (dot product) on two 2D matrices.
# Inputs:
# a = np.array([[1, 2], [3, 4]])
# b = np.array([[5, 6], [7, 8]])
# Expected Output:
# [[19 22]
#  [43 50]]

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
np.dot(a,b)
```

```python
# Problem: Find the sum of all elements in the array.
# Input: a = np.array([[1, 2], [3, 4]])
# Expected Output: 10

a = np.array([[1, 2], [3, 4]])
np.sum(a)
```

```python
# Problem: Calculate the sum of each column.
# Inputs:
# a = np.array([[1, 2, 3],
#               [4, 5, 6]])
# Expected Output: [5 7 9]

a = np.array([[1, 2, 3],[4, 5, 6]])
np.sum(a,axis=0)
```

```python
# Problem: Calculate the average value of each row .
# Input:
# a = np.array([[10, 20, 30],
#               [40, 50, 60]])
# Expected Output: [20. 50.]

a = np.array([[10, 20, 30],[40, 50, 60]])
np.mean(a,axis=1)
```

```python
# Problem: Find the smallest value in the entire array.
# Input: a = np.array([10, 5, 8, 3, 9])
# Expected Output: 3

a = np.array([10, 5, 8, 3, 9])
int(np.min(a))
```

```python
# Problem: Find the index (position) of the maximum value in the array.
# Input: a = np.array([10, 50, 30, 20])
# Expected Output: 1 (Since 50 is at index 1)

a = np.array([10, 50, 30, 20])
int(np.argmax(a))
```

```python
# Problem: Find the index of the maximum value in each row.
# Input:
# a = np.array([[1, 2, 3],
#               [6, 5, 4]])
# Expected Output: [2 0] (3 is max in row 0 at idx 2; 6 is max in row 1 at idx 0)

a = np.array([[1, 2, 3],[6, 5, 4]])
np.argmax(a,axis=1)
```

```python
# Problem: Compute the cumulative sum of the elements.
# Input: a = np.array([1, 2, 3, 4, 5])
# Expected Output: [ 1 3 6 10 15]

a = np.array([1, 2, 3, 4, 5])
np.cumsum(a)
```

```python
# Problem: Calculate the difference between the maximum and minimum values (ptp).
# Input: a = np.array([1, 5, 10, 20])
# Expected Output: 19

a = np.array([1, 5, 10, 20])
int(np.max(a))-int(np.min(a))
```

```python
# Problem: Add a 1D array b to each row of a 2D array a.
# Input:
# a = np.array([[1, 1, 1],[1, 1, 1]])
# b = np.array([0, 1, 2])
# Expected Output:[[1 2 3],[1 2 3]]

a = np.array([[1, 1, 1],[1, 1, 1]])
b = np.array([0, 1, 2])
np.add(a,b)
```

```python
# Problem: Multiply a 2D array a by a column vector b.
# Input:
# a = np.array([[1, 2, 3],
#               [4, 5, 6]])
# b = np.array([[10],
#               [20]])
# Expected Output:[[ 10  20  30], [ 80 100 120]]

a = np.array([[1, 2, 3],[4, 5, 6]])
b = np.array([[10],[20]])
a*b
```

```python
# Problem: Limit the values in the array so that values less than 10 become 10, and values greater than 20 become 20.
# Input: a = np.array([5, 10, 15, 20, 25])
# Expected Output: [10 10 15 20 20]

a = np.array([5, 10, 15, 20, 25])
np.clip(a,10,20)
```

```python
# Problem: Calculate the Euclidean distance between two points (vectors) a and b.(Hint: $\sqrt{\sum(a_i - b_i)^2}$)
# Input:a = np.array([1, 2])
# b = np.array([4, 6])
# Expected Output: 5.0

a = np.array([1, 2])
b = np.array([4, 6])
# np.sqrt(np.sum((a-b)**2))  or
np.linalg.norm(a - b)
```

```python
# Problem: Normalize array a by subtracting the mean and dividing by the standard deviation.
# Input: a = np.array([10, 20, 30, 40, 50])
# Expected Output: [-1.414 -0.707 0. 0.707 1.414] (approx)

a = np.array([10, 20, 30, 40, 50])
np.round((a-np.mean(a))/np.std(a),3)
```
